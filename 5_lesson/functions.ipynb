–ü—Ä–∏–º–µ—Ä –¥–ª—è C++: https://repl.it/@vbugaevskii/ValueVsRefernceCpp

–ü—Ä–∏–º–µ—Ä –¥–ª—è Java: https://repl.it/@vbugaevskii/ValueVsRefernceJava

–ö–∞–∫ –≤–æ–æ–±—Ä–∞–∑–∏—Ç—å —Å–µ–±–µ –ø–µ—Ä–µ–¥–∞—á—É –ø–æ —Å—Å—ã–ª–∫–µ, –µ—Å–ª–∏ —è –µ—â–µ –Ω–µ –±—ã–ª –∑–Ω–∞–∫–æ–º —Å Java –∏ Python, –Ω–æ –∑–Ω–∞–∫–æ–º —Å C++?

1. –°–æ–∑–¥–∞–µ—Ç—Å—è —Å—Å—ã–ª–∫–∞ `ref_A` –Ω–∞ –æ–±—ä–µ–∫—Ç `obj_A`;
2. –°—Å—ã–ª–∫–∞ `ref_A` –ø–µ—Ä–µ–¥–∞–µ—Ç—Å—è –≤ —Ñ—É–Ω–∫—Ü–∏—é –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é, —Ç.–µ. —Å–æ–∑–¥–∞–µ—Ç—Å—è –∫–æ–ø–∏—è `copy_ref_A`.

–ü–æ—ç—Ç–æ–º—É —á—Ç–æ–±—ã –∏–∑–º–µ–Ω–∏—Ç—å –æ–±—ä–µ–∫—Ç `obj_A` –ø–æ —Å—Å—ã–ª–∫–µ `copy_ref_A`, –Ω—É–∂–Ω–æ –≤—ã–∑–≤–∞—Ç—å –º–µ—Ç–æ–¥ –º–µ–Ω—è—é—â–∏–π —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ `obj_A`, –ª–∏–±–æ "—è–≤–Ω–æ –æ–±—Ä–∞—Ç–∏—Ç—å—Å—è" –∫ –¥–∞–Ω–Ω—ã–º `obj_A` —á–µ—Ä–µ–∑ —Å—Å—ã–ª–∫—É `cop_ref_A`.

`copy_ref_A = copy_ref_B` –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ `copy_ref_A` —Ç–µ–ø–µ—Ä—å —Å–æ–≤–ø–∞–¥–∞–µ—Ç c `copy_ref_B` –∏ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –æ–±—ä–µ–∫—Ç B.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º —Å—Å—ã–ª–∫–∏ –æ—á–µ–Ω—å –ø–æ—Ö–æ–∂–∏ –Ω–∞ **—É–∫–∞–∑–∞—Ç–µ–ª–∏** –≤ C++.

````python
result = "GLOBAL"

def func():
    global result     # —Ä–∞–∑—Ä–µ—à–∞–µ—Ç –º–µ–Ω—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –≥–ª–æ–±–∞–ª—å–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
    result = "LOCAL"
    print("[local]\t\t", result)

print("[global]\t", result)
func()
print("[global]\t", result)
````

````python
result = 'GLOBAL'

def func_outer():  # outer function
    result = 'ENCLOSED'
    print("[enclosed]\t", result)

    def func():    # nested (inner) function
        result = 'LOCAL'
        print("[local]\t\t", result)

    func()
    print("[enclosed]\t", result)

print("[global]\t", result)
func_outer()
print("[global]\t", result)
````

````python
result = 'GLOBAL'

def func_outer():
    result = 'ENCLOSED'
    print("[enclosed]\t", result)
    
    def func():
        global result
        result = 'LOCAL'
        print("[local]\t\t", result)
        
    func()
    print("[enclosed]\t", result)

print("[global]\t", result)
func_outer()
print("[global]\t", result)
````

````python
result = 'GLOBAL'

def func_outer():
    result = 'ENCLOSED'
    print("[enclosed]\t", result)
    
    def func():
        nonlocal result   # —Ä–∞–∑—Ä–µ—à–∞–µ—Ç –º–µ–Ω—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ enclosed –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
        result = 'LOCAL'
        print("[local]\t\t", result)
        
    func()
    print("[enclosed]\t", result)

print("[global]\t", result)
func_outer()
print("[global]\t", result)
````

````python
result = 'GLOBAL'

def func_outer():
    result = 'ENCLOSED'
    
    def func():
        nonlocal result   # —Ä–∞–∑—Ä–µ—à–∞–µ—Ç –º–µ–Ω—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ enclosed –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
        result = 'LOCAL'
    
    print("closure:", func.__closure__)
    print("closure content:", func.__closure__[0].cell_contents)
    
    func()

print("[global]\t", result)
func_outer()
print("[global]\t", result)
````

## –§—É–Ω–∫—Ü–∏—è reduce

````python
from functools import reduce
````

````python
def my_reduce(func, seq):
    res = seq[0]
    for elem in seq[1:]:
        res = func(res, elem)
    return res
````

````python
from time import sleep

def cached(func):
    cache = dict()
    @functools.wraps(func)
    def wrapper(*args):
        key = (func, args)
        if key not in cache:
            cache[key] = func(*args)
        return cache[key]
    return wrapper

@cached
def power2(x):
    sleep(3)
    return 2 ** x

print(power2(8))
print(power2(8))
print(power2(4))
print(power2(8))
print(power2(4))
````

````python
from functools import lru_cache

@lru_cache(maxsize=5)
def power2(x):
    sleep(3)
    return 2 ** x

print(power2(8))
print(power2(8))
print(power2(4))
print(power2(8))
print(power2(4))
````
